```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar el dataset
df = pd.read_csv('C:/git1/Grocery-Sales-Analysis/data/processed/Supermart processed data.csv')
```

```python
# Análisis descriptivo
descriptive_stats = df.describe()
print(descriptive_stats)

# Para columnas categóricas
categorical_summary = df.describe(include='object')
print(categorical_summary)
```

```python
# Histograma de Ventas
plt.figure(figsize=(10, 6))
sns.histplot(df['Sales'], bins=30, kde=True)
plt.title('Distribución de Ventas')
plt.xlabel('Ventas')
plt.ylabel('Frecuencia')
plt.show()

# Gráfico de barras de categorías
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='Category')
plt.title('Conteo de Categorías')
plt.xlabel('Categoría')
plt.ylabel('Conteo')
plt.xticks(rotation=45)
plt.show()
```

```python
# Boxplot de Ventas
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['Sales'])
plt.title('Boxplot de Ventas')
plt.xlabel('Ventas')
plt.show()
```

```python
# Matriz de correlación
correlation_matrix = df[['Sales', 'Discount', 'Profit']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm', square=True)
plt.title('Matriz de Correlación')
plt.show()
```

```python
# Gráfico de barras de ventas por región
plt.figure(figsize=(12, 6))
sns.barplot(data=df, x='Region', y='Sales', estimator=sum)
plt.title('Ventas Totales por Región')
plt.xlabel('Región')
plt.ylabel('Ventas Totales')
plt.xticks(rotation=45)
plt.show()
```
